```python
# ================================================================
# === CELDA 3: CLASES Y FUNCIONES DE LA SOLUCIÓN DQN ===
# ================================================================

# ----------------------------------------------------------------
# CONFIGURACIÓN DEL ESTADO DE ENTRADA (PREPROCESAMIENTO ATARI)
# ----------------------------------------------------------------
# Las observaciones del entorno Enduro-v0 se preprocesan siguiendo
# la práctica habitual en entornos Atari bajo ALE:
# - Cada frame se convierte a escala de grises.
# - Se redimensiona a 84×84 píxeles.
#
# Adicionalmente, se emplea "frame stacking" (apilado de 4 frames),
# lo que permite incorporar información temporal y dinámica del entorno
# (p. ej., desplazamiento y movimiento relativo).
#
# Referencia:
# https://ale.farama.org/environments/enduro/
HEIGHT, WIDTH = 84, 84
FRAME_STACK = 4  # Número de frames apilados (estado)
INPUT_SHAPE = (HEIGHT, WIDTH, FRAME_STACK)

# ----------------------------------------------------------------
# 3.1 AtariProcessor
# ----------------------------------------------------------------
class AtariProcessor:
    """Preprocesador de observaciones Atari: redimensiona, convierte a gris y retorna uint8."""
    def __init__(self, input_shape=(84, 84)):
        # PIL recibe (ancho, alto). Si es cuadrado, el orden no afecta.
        self.input_shape = input_shape

    def process_observation(self, observation):
        """Convierte un frame RGB (210x160x3) a 84x84 en escala de grises (uint8)."""
        img = Image.fromarray(observation)
        img = img.resize(self.input_shape).convert('L')
        processed_observation = np.array(img)
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        """Asegura tipo float32. La normalización /255 se realiza externamente."""
        return batch.astype('float32')

    def process_reward(self, reward):
        """Aplica clipping de recompensa típico en Atari (rango [-1, 1])."""
        return np.clip(reward, -1., 1.)

# ----------------------------------------------------------------
# 3.2 Modelo DQN (Red Neuronal)
# ----------------------------------------------------------------
# Red neuronal profunda que aproxima la función Q(s, a).
# Se utiliza una arquitectura convolucional inspirada en el DQN original
# (Mnih et al.), adecuada para estados visuales representados como
# stacks de frames.
def build_model(input_shape, num_actions):
    """
    Construye el modelo Q(s, a) basado en una red convolucional (CNN).

    La red recibe como entrada un estado compuesto por varios frames apilados
    y devuelve un valor Q por cada acción posible, implementando un
    DQN clásico (“Natural”), sin Double DQN, Dueling ni Prioritized Replay.
    """
    model = Sequential()

    # Primera capa convolucional: extracción de características espaciales gruesas
    model.add(Conv2D(
        32, (8, 8),
        strides=(4, 4),
        activation='relu',
        input_shape=input_shape
    ))

    # Segunda capa convolucional: captura patrones visuales más complejos
    model.add(Conv2D(
        64, (4, 4),
        strides=(2, 2),
        activation='relu'
    ))

    # Tercera capa convolucional: refinamiento de características de alto nivel
    model.add(Conv2D(
        64, (3, 3),
        strides=(1, 1),
        activation='relu'
    ))

    # Aplanado para conectar con capas densas
    model.add(Flatten())

    # Capa densa: combinación no lineal de características
    model.add(Dense(512, activation='relu'))

    # Capa de salida: un valor Q(s, a) por cada acción discreta
    model.add(Dense(num_actions, activation='linear'))
    return model

# ----------------------------------------------------------------
# 3.3 Replay Buffer
# ----------------------------------------------------------------
# Buffer de repetición de experiencias para romper correlación temporal
# y estabilizar el aprendizaje, muestreando minibatches aleatorios.
class ReplayBuffer:
    """
    Buffer circular que almacena transiciones:
    (estado, acción, recompensa, siguiente_estado, done)
    (s, a, r, s', done).

    Permite muestrear minibatches aleatorios para el entrenamiento.
    """
    def __init__(self, capacity):
        # Cola circular con capacidad máxima fija
        self.memory = deque(maxlen=capacity)

    def add(self, state, action, reward, next_state, done):
        # Almacena una transición; al llenarse, descarta automáticamente la más antigua
        self.memory.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        # Retorna un minibatch aleatorio si hay suficientes muestras
        if len(self.memory) < batch_size:
            return None
        return random.sample(self.memory, batch_size)

    def __len__(self):
        # Número actual de transiciones almacenadas
        return len(self.memory)

# ----------------------------------------------------------------
# 3.4 DQNAgent (Agente Principal)
# ----------------------------------------------------------------
# Encapsula la lógica completa del agente DQN:
# - Selección de acciones (ε-greedy)
# - Almacenamiento de experiencias
# - Redes policy y target
# - Entrenamiento mediante minimización del error TD
class DQNAgent:
    def __init__(self, state_shape, num_actions, buffer_capacity, gamma, lr, update_target_steps):
        # Dimensiones del estado y número de acciones
        self.state_shape = state_shape
        self.num_actions = num_actions

        # Factor de descuento (ecuación de Bellman)
        self.gamma = gamma

        # Frecuencia de actualización de la red objetivo (target)
        self.update_target_steps = update_target_steps

        # Contador de actualizaciones (para sincronizar la red target)
        self.step_count = 0

        # Preprocesador de observaciones Atari
        self.processor = AtariProcessor(input_shape=(WIDTH, HEIGHT))

        # Red principal (policy network)
        self.policy_net = build_model(state_shape, num_actions)

        # Red objetivo (target network)
        self.target_net = build_model(state_shape, num_actions)

        # Optimizador Adam (configuración típica en DQN)
        self.optimizer = Adam(learning_rate=lr, epsilon=1e-4)

        # Pérdida Huber (robusta frente a grandes errores TD)
        self.loss_fn = Huber(delta=1.0)

        # Compilación de la red principal (la red target no requiere compilación)
        self.policy_net.compile(optimizer=self.optimizer, loss=self.loss_fn)

        # Sincronización inicial de pesos: target ← policy
        self.target_net.set_weights(self.policy_net.get_weights())

        # Replay buffer
        self.memory = ReplayBuffer(buffer_capacity)

    def select_action(self, state, epsilon):
        """
        Selección de acción con política ε-greedy:
        - Con probabilidad ε: exploración (acción aleatoria).
        - Con probabilidad 1−ε: explotación (argmax Q).
        """
        # Exploración: acción aleatoria
        if random.random() < epsilon:
            return random.randrange(self.num_actions)

        # Normalización del estado a [0, 1] para la inferencia
        state_in = self.processor.process_state_batch(
            np.expand_dims(state, 0)
        ) / 255.0

        # Inferencia con la red principal
        q_values = self.policy_net(state_in, training=False)

        # Selección de la acción con mayor valor Q
        return int(tf.argmax(q_values[0]).numpy())


        # Decorador de TensorFlow que compila la función en grafo.
        # Esto reduce el overhead de Python y acelera el paso de entrenamiento
        # en CPU/GPU.
        @tf.function
        def train_step(self, states, actions, rewards, next_states, dones):
            """
            Ejecuta un paso de entrenamiento del DQN clásico.

            A partir de un minibatch (s, a, r, s', done), el procedimiento:
            - Calcula Q(s, ·) usando la red de política.
            - Extrae Q(s, a) correspondiente a la acción tomada.
            - Calcula el objetivo con la red target: r + γ max_a' Q(s', a') (si no terminal).
            - Minimiza el error TD actualizando únicamente la red principal.
            """
            # Registro automático de gradientes
            with tf.GradientTape() as tape:

                # Q(s, ·) con la red principal
                current_q_values = self.policy_net(states, training=True)

                # Extracción de Q(s, a) mediante máscara one-hot
                action_mask = tf.one_hot(actions, self.num_actions, dtype=tf.float32)
                predicted_q = tf.reduce_sum(current_q_values * action_mask, axis=1)

                # Q(s', ·) con la red target
                next_q_values = self.target_net(next_states, training=False)

                # max_a' Q(s', a')
                max_next_q = tf.reduce_max(next_q_values, axis=1)

                # Objetivo TD: r + γ max_a' Q(s', a') · (1 − done)
                dones_f = tf.cast(dones, tf.float32)
                target_q = rewards + self.gamma * max_next_q * (1.0 - dones_f)

                # Pérdida TD (Huber)
                loss = self.loss_fn(target_q, predicted_q)

            # Gradientes respecto a la red principal
            grads = tape.gradient(loss, self.policy_net.trainable_variables)

            # Actualización de pesos (solo policy_net)
            self.optimizer.apply_gradients(
                zip(grads, self.policy_net.trainable_variables)
            )

            # Retorna la pérdida para monitoreo
            return loss

    # ----------------------------------------------------------------
    # MÉTODO DE OPTIMIZACIÓN / ENTRENAMIENTO
    # ----------------------------------------------------------------
    # Coordina el muestreo desde el replay buffer, el preprocesamiento
    # de datos y la ejecución del paso de entrenamiento. Además, gestiona
    # la sincronización periódica de la red target.
    def optimize_model(self, batch_size):
        """
        Muestrea un minibatch del replay buffer, preprocesa los datos
        y ejecuta un paso de entrenamiento del DQN.

        También sincroniza periódicamente la red target con la red de
        política cada N actualizaciones para estabilizar el aprendizaje.
        """
        # Muestreo aleatorio de experiencias
        batch = self.memory.sample(batch_size)
        if batch is None:
            # Si no hay suficientes muestras aún, no se entrena
            return 0.0

        # Desempaquetado de transiciones (s, a, r, s', done)
        states, actions, rewards, next_states, dones = zip(*batch)

        # Conversión a NumPy y normalización a [0, 1]
        states = np.asarray(states, dtype=np.float32) / 255.0
        next_states = np.asarray(next_states, dtype=np.float32) / 255.0
        rewards = np.asarray(rewards, dtype=np.float32)
        actions = np.asarray(actions, dtype=np.int32)
        dones = np.asarray(dones, dtype=np.bool_)

        # Paso de entrenamiento
        loss = self.train_step(states, actions, rewards, next_states, dones)

        # Contador de actualizaciones
        self.step_count += 1

        # Sincronización periódica de la red target
        if self.step_count % self.update_target_steps == 0:
            self.target_net.set_weights(self.policy_net.get_weights())

        return float(loss.numpy())

    # ----------------------------------------------------------------
    # CARGA DE PESOS
    # ----------------------------------------------------------------
    def load_weights(self, filename):
        """
        Carga pesos guardados en la red de política y sincroniza la red target.
        """
        self.policy_net.load_weights(filename)
        self.target_net.set_weights(self.policy_net.get_weights())

    # ----------------------------------------------------------------
    # GUARDADO DE PESOS
    # ----------------------------------------------------------------
    def save_weights(self, filename):
        """
        Guarda los pesos actuales de la red de política.
        Se recomienda el sufijo '.weights.h5' por convención de Keras.
        """
        self.policy_net.save_weights(filename, overwrite=True)

# ----------------------------------------------------------------
# FUNCIÓN AUXILIAR: ACTUALIZACIÓN DEL FRAME STACK
# ----------------------------------------------------------------
def update_state_stack(old_state, new_frame):
    """
    Actualiza el estado apilado de frames (frame stack):
    - Desplaza el stack una posición.
    - Inserta el nuevo frame procesado al final.
    """
    new_state = np.roll(old_state, shift=-1, axis=2)
    new_state[:, :, -1] = new_frame
    return new_state

# ----------------------------------------------------------------
# FUNCIÓN DE EVALUACIÓN DEL AGENTE
# ----------------------------------------------------------------
def evaluate_agent(agent, env, num_episodes=100, epsilon=0.05):
    """
    Evalúa el agente entrenado durante un número fijo de episodios,
    utilizando una política ε-greedy con exploración baja.
    """
    print(f"\nCorriendo {num_episodes} episodios de prueba (epsilon={epsilon})...")
    test_rewards = []

    for _ in range(num_episodes):
        # Reset compatible con distintas versiones de Gym
        reset_out = env.reset()
        obs = reset_out[0] if isinstance(reset_out, tuple) else reset_out

        # Estado inicial: frame procesado repetido FRAME_STACK veces
        proc_frame = agent.processor.process_observation(obs)
        state = np.stack([proc_frame] * FRAME_STACK, axis=-1)

        done = False
        ep_reward = 0.0

        while not done:
            # Acción con política ε-greedy
            action = agent.select_action(state, epsilon=epsilon)

            step_out = env.step(action)

            # Compatibilidad con Gym moderno y versiones antiguas
            if len(step_out) == 5:
                next_obs, reward, terminated, truncated, info = step_out
                done = bool(terminated or truncated)
            else:
                next_obs, reward, done, info = step_out

            ep_reward += reward

            # Actualización del estado apilado
            proc_next_frame = agent.processor.process_observation(next_obs)
            state = update_state_stack(state, proc_next_frame)

        test_rewards.append(ep_reward)

    avg_test_reward = float(np.mean(test_rewards))
    std_test_reward = float(np.std(test_rewards))

    print(f"\n--- Resultado Final del Test ({num_episodes} episodios) ---")
    print(f"Recompensa Media: {avg_test_reward:.2f}")
    print(f"Desviación Estándar: {std_test_reward:.2f}")

    return avg_test_reward
```

```python
# ============================================================
# CELDA 04: DQN COMPLETO PARA ENDuro-v0 (gym 0.17.3 + atari_py)
# - Corrige "loss=0" usando terminal real (game over), no pérdida de vida.
# ============================================================

import os, random
import numpy as np
from collections import deque

import gym
import cv2
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam

# -------------------------
# CONFIG
# -------------------------
ENV_NAME = "Enduro-v0"

HEIGHT, WIDTH = 84, 84
FRAME_STACK = 4
INPUT_SHAPE = (HEIGHT, WIDTH, FRAME_STACK)

GAMMA = 0.99
LEARNING_RATE = 2.5e-4
BUFFER_CAPACITY = 200_000
BATCH_SIZE = 32
WARMUP_STEPS = 50_000       # antes de entrenar
TRAIN_EVERY = 4
TARGET_UPDATE_STEPS = 10_000

NB_STEPS = start_step + 20000

EPS_MAX = 1.0
EPS_MIN = 0.1
EPS_DECAY_STEPS = 1_000_000

CHECKPOINT_PATH = "/content/drive/My Drive/08_MIAR/actividades/proyecto practicoG/dqn_enduro_checkpoint.weights.h5"
PROGRESS_FILE   = "/content/drive/My Drive/08_MIAR/actividades/proyecto practicoG/dqn_enduro_progress.txt"
LOG_PATH = "/content/drive/My Drive/08_MIAR/actividades/proyecto practicoG/historial_entrenamiento.txt"

os.makedirs(os.path.dirname(LOG_PATH), exist_ok=True)

# -------------------------
# UTILIDADES
# -------------------------
def preprocess_frame(obs):
    """Frame RGB -> gris 84x84 uint8."""
    gray = cv2.cvtColor(obs, cv2.COLOR_RGB2GRAY)
    resized = cv2.resize(gray, (WIDTH, HEIGHT), interpolation=cv2.INTER_AREA)
    return resized.astype(np.uint8)

def update_state_stack(state, new_frame):
    """state: (84,84,4), new_frame: (84,84) -> (84,84,4)"""
    return np.concatenate([state[..., 1:], new_frame[..., None]], axis=-1)

# -------------------------
# REPLAY BUFFER
# -------------------------
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def add(self, s, a, r, s2, terminal):
        self.buffer.append((s, a, r, s2, terminal))

    def sample(self, batch_size):
        idx = np.random.choice(len(self.buffer), batch_size, replace=False)
        batch = [self.buffer[i] for i in idx]
        s, a, r, s2, d = map(np.array, zip(*batch))
        return s, a, r, s2, d

    def __len__(self):
        return len(self.buffer)

# -------------------------
# MODELO (CNN Atari clásico)
# -------------------------
def build_q_network(input_shape, num_actions):
    inp = layers.Input(shape=input_shape)
    x = layers.Lambda(lambda t: tf.cast(t, tf.float32) / 255.0)(inp)

    x = layers.Conv2D(32, 8, strides=4, activation="relu")(x)
    x = layers.Conv2D(64, 4, strides=2, activation="relu")(x)
    x = layers.Conv2D(64, 3, strides=1, activation="relu")(x)
    x = layers.Flatten()(x)
    x = layers.Dense(512, activation="relu")(x)
    out = layers.Dense(num_actions)(x)

    return tf.keras.Model(inp, out)

# -------------------------
# AGENTE DQN
# -------------------------
class DQNAgent:
    def __init__(self, state_shape, num_actions, buffer_capacity, gamma, lr, update_target_steps):
        self.num_actions = num_actions
        self.gamma = gamma
        self.update_target_steps = update_target_steps
        self.train_updates = 0

        self.memory = ReplayBuffer(buffer_capacity)

        self.policy_net = build_q_network(state_shape, num_actions)
        self.target_net = build_q_network(state_shape, num_actions)
        self.target_net.set_weights(self.policy_net.get_weights())

        self.optimizer = Adam(learning_rate=lr)

    def save_weights(self, path):
        self.policy_net.save_weights(path)

    def load_weights(self, path):
        self.policy_net.load_weights(path)
        self.target_net.set_weights(self.policy_net.get_weights())

    def select_action(self, state, epsilon):
        if random.random() < epsilon:
            return random.randrange(self.num_actions)
        state_b = np.expand_dims(state, axis=0)  # (1,84,84,4)
        q = self.policy_net(state_b, training=False)[0].numpy()
        return int(np.argmax(q))

    @tf.function
    def train_step(self, states, actions, rewards, next_states, terminals):
        actions  = tf.cast(actions, tf.int32)
        rewards  = tf.cast(rewards, tf.float32)
        terminals = tf.cast(terminals, tf.float32)

        next_q = self.target_net(next_states, training=False)
        max_next_q = tf.reduce_max(next_q, axis=1)

        # Si terminal==1 (game over), no bootstrap
        target = rewards + (1.0 - terminals) * self.gamma * max_next_q

        with tf.GradientTape() as tape:
            q_values = self.policy_net(states, training=True)
            mask = tf.one_hot(actions, self.num_actions)
            q_action = tf.reduce_sum(q_values * mask, axis=1)
            loss = tf.reduce_mean(tf.square(target - q_action))

        grads = tape.gradient(loss, self.policy_net.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.policy_net.trainable_variables))
        return loss

    def optimize_model(self, batch_size):
        if len(self.memory) < batch_size:
            return 0.0

        s, a, r, s2, terminal = self.memory.sample(batch_size)

        s  = s.astype(np.uint8)
        s2 = s2.astype(np.uint8)
        a  = a.astype(np.int32)
        r  = r.astype(np.float32)
        terminal = terminal.astype(np.float32)

        loss = self.train_step(s, a, r, s2, terminal)

        self.train_updates += 1
        if self.train_updates % self.update_target_steps == 0:
            self.target_net.set_weights(self.policy_net.get_weights())

        return float(loss.numpy())

# -------------------------
# ENTORNO + AGENTE
# -------------------------
env = gym.make(ENV_NAME)
num_actions = env.action_space.n

agent = DQNAgent(
    state_shape=INPUT_SHAPE,
    num_actions=num_actions,
    buffer_capacity=BUFFER_CAPACITY,
    gamma=GAMMA,
    lr=LEARNING_RATE,
    update_target_steps=TARGET_UPDATE_STEPS
)

# -------------------------
# REANUDACIÓN
# -------------------------
start_step = 0
if os.path.exists(CHECKPOINT_PATH):
    print(f"Checkpoint encontrado: Cargando pesos desde {CHECKPOINT_PATH}...")
    agent.load_weights(CHECKPOINT_PATH)

if os.path.exists(PROGRESS_FILE):
    with open(PROGRESS_FILE, "r") as f:
        try:
            start_step = int(f.read().strip())
        except:
            start_step = 0

print(f"Reanudando entrenamiento desde el paso: {start_step}")

# -------------------------
# TRAIN LOOP
# -------------------------
episode_rewards = deque(maxlen=100)
all_episode_rewards = []

total_steps = start_step
episode_count = 0
last_loss = 0.0

print(f"Iniciando entrenamiento en {ENV_NAME} (Objetivo: {NB_STEPS} pasos)...")

while total_steps < NB_STEPS:
    obs = env.reset()
    frame = preprocess_frame(obs)
    state = np.stack([frame]*FRAME_STACK, axis=-1)

    done = False
    ep_reward = 0.0

    # vidas actuales (para detectar pérdida de vida)
    prev_lives = env.ale.lives()

    while not done:
        total_steps += 1

        if total_steps < WARMUP_STEPS:
            epsilon = EPS_MAX
        else:
            steps_decayed = total_steps - WARMUP_STEPS
            epsilon = max(EPS_MIN, EPS_MAX - (EPS_MAX - EPS_MIN) * (steps_decayed / EPS_DECAY_STEPS))

        action = agent.select_action(state, epsilon)
        next_obs, reward, done, info = env.step(action)

        next_frame = preprocess_frame(next_obs)
        next_state = update_state_stack(state, next_frame)

        # Reward clipping clásico (no lo vuelve siempre 0)
        clipped_r = np.sign(reward).astype(np.float32)

        # ---------------------------
        # CLAVE: terminal REAL = game over (vidas == 0)
        # No usar done por pérdida de vida como terminal.
        # ---------------------------
        lives = env.ale.lives()
        terminal = done and (lives == 0)

        agent.memory.add(state, action, clipped_r, next_state, terminal)

        # Si solo se perdió una vida, consideramos "fin de episodio" para tu loop,
        # pero NO terminal para el TD target.
        # (esto mantiene tu while por done, pero TD aprende bien)
        prev_lives = lives

        state = next_state
        ep_reward += reward

        if total_steps > WARMUP_STEPS and total_steps % TRAIN_EVERY == 0:
            last_loss = agent.optimize_model(BATCH_SIZE)

        if total_steps % 10_000 == 0:
            print(
                f"Step: {total_steps} | Episodes: {episode_count} | "
                f"Avg Rwd (100): {np.mean(episode_rewards) if len(episode_rewards)>0 else 0.0:.2f} | "
                f"Loss: {last_loss:.8f} | Epsilon: {epsilon:.4f} | Buffer: {len(agent.memory)}"
            )

        if total_steps % 50_000 == 0:
            agent.save_weights(CHECKPOINT_PATH)
            with open(PROGRESS_FILE, "w") as f:
                f.write(str(total_steps))
            print(f"\nCHECKPOINT GUARDADO: Paso {total_steps} (Peso y Progreso)\n")

    episode_rewards.append(ep_reward)
    all_episode_rewards.append(ep_reward)
    episode_count += 1
```

```python
def run_episode(agent, env, epsilon=0.05, visualize=False, repeat=1, max_steps_ep=1_000_000):
    obs = env.reset()
    frame = preprocess_frame(obs)
    state = np.stack([frame]*FRAME_STACK, axis=-1)

    done = False
    ep_r = 0.0
    steps_ep = 0

    while not done and steps_ep < max_steps_ep:
        steps_ep += 1
        action = agent.select_action(state, epsilon=epsilon)

        total_r = 0.0
        for _ in range(repeat):
            next_obs, r, done, info = env.step(action)
            total_r += float(r)
            if done:
                break

        state = update_state_stack(state, preprocess_frame(next_obs))
        ep_r += total_r

        if visualize:
            env.render()

    return ep_r

def check_100_consecutive(agent, threshold=200.0, max_episodes=5000, epsilon=0.00, repeat=1, max_steps_ep=1_000_000):
    env = gym.make(ENV_NAME)
    streak = 0
    best = 0

    for i in range(1, max_episodes + 1):
        r = run_episode(agent, env, epsilon=epsilon, visualize=False, repeat=repeat, max_steps_ep=max_steps_ep)
        passed = (r >= threshold)

        if not passed and streak > 0:
            print(f"❌ Se rompió la racha en ep {i}: R={r:.2f} < {threshold} (racha previa={streak})")

        streak = streak + 1 if passed else 0
        best = max(best, streak)

        # >>> IMPRIMIR TODOS LOS EPISODIOS
        print(f"Ep {i:4d} | R={r:7.2f} | passed={passed} | streak={streak} | best={best}")

        if streak >= 100:
            env.close()
            print(f"\n✅ Conseguido: 100 episodios consecutivos con R >= {threshold}")
            return True

    env.close()

check_100_consecutive(agent, threshold=200.0, max_episodes=2000, epsilon=0.0, repeat=1, max_steps_ep=1_000_000)
```

## JUSTIFICACIÓN DE LOS PARÁMETROS SELECCIONADOS Y DE LOS RESULTADOS OBTENIDOS

En el desarrollo del presente proyecto se implementó un agente basado en Deep Q-Learning (DQN) para el entorno Enduro-v0, perteneciente al conjunto de entornos Atari. La selección de los hiperparámetros se realizó considerando las recomendaciones establecidas en la literatura sobre aprendizaje por refuerzo profundo, así como la experiencia obtenida durante las pruebas preliminares del modelo.


## ELECCIÓN Y JUSTIFICACIÓN DE LOS PARÁMETROS

En primer lugar, se estableció un factor de descuento (γ = 0.99) con el propósito de priorizar las recompensas futuras dentro del proceso de aprendizaje. Este valor es ampliamente utilizado en entornos de tipo Atari, ya que permite que el agente tenga en cuenta las consecuencias de sus acciones a largo plazo, favoreciendo estrategias sostenidas durante el desarrollo del juego.

Asimismo, se definió una tasa de aprendizaje de 2.5 × 10⁻⁴, valor que proporciona un equilibrio adecuado entre estabilidad y velocidad de convergencia. Una tasa de aprendizaje elevada podría provocar oscilaciones en la función de pérdida, mientras que valores excesivamente bajos ralentizarían considerablemente el proceso de entrenamiento.

El Replay Buffer se configuró con una capacidad de 200 000 transiciones, lo cual permite almacenar una cantidad suficiente de experiencias provenientes de diferentes estados del entorno. Este mecanismo resulta fundamental para romper la correlación temporal entre muestras consecutivas, mejorando la estabilidad del entrenamiento y evitando que el agente aprenda patrones demasiado dependientes de secuencias específicas de estados.

Por otra parte, el tamaño de lote (batch size) se fijó en 32 muestras, valor comúnmente utilizado en algoritmos DQN debido a que permite mantener un buen balance entre estabilidad del gradiente y eficiencia computacional.

El entrenamiento se inició luego de 50 000 pasos de calentamiento (warm-up). Durante esta etapa el agente interactúa con el entorno sin actualizar los parámetros de la red, permitiendo llenar el buffer de experiencias con datos iniciales diversos. Esta práctica evita que el proceso de aprendizaje se base en un conjunto de muestras demasiado reducido o altamente correlacionado.

En relación con la estabilidad del algoritmo, se empleó una red objetivo (target network) que se actualiza cada 10 000 pasos, mecanismo que reduce el problema conocido como moving target problem, característico de los métodos de aproximación de valores en aprendizaje por refuerzo profundo.

Para la estrategia de exploración se utilizó el enfoque ε-greedy, iniciando con ε = 1.0 para fomentar la exploración completa del entorno durante las primeras fases del entrenamiento. Posteriormente, este valor se reduce progresivamente hasta alcanzar ε = 0.1, permitiendo que el agente explote el conocimiento adquirido mientras mantiene un nivel mínimo de exploración que evita caer en políticas subóptimas.

Finalmente, se aplicó un proceso de preprocesamiento de las observaciones, reduciendo las imágenes del entorno a una resolución de 84 × 84 píxeles en escala de grises, además de utilizar un apilamiento de cuatro fotogramas consecutivos (frame stacking). Este procedimiento permite al modelo capturar información temporal relevante, como la velocidad y dirección de los objetos en el entorno.
